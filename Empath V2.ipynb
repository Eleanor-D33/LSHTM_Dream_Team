```python
# defining function to analyse sentences 
# can be altered to add category and normalise values 

def empath_analyse(sentence):
    x = lexicon.analyze(sentence) #,categories=["violence"])
    return x
```

```python
# reading tweet data 

tweets = pd.read_csv('/Users/eleanordavies/Desktop/demo_tweets_clean.csv')
### tweets["id"] = tweets.index
tweets.head()
```

## Running tweets through empath 
- Create an array of the 'cleaned' text from the tweets 
- Run a for loop to analyse tweets usinh emapth_analyse function 
- Convert data dictionary into dataframe  
- Append data dictionary dataframe to the exsiting tweet dataframe 

```python
data_to_list = tweets['selected_tweets'].values.tolist()
data_to_list = np.array(data_to_list)

data =[]
for i in range(len(data_to_list)):
    data.append(empath_analyse(data_to_list[i])),
data_dict = pd.DataFrame.from_dict(data)
data_dict.head()
```

```python
result = pd.concat([tweets, data_dict], axis=1, join="inner")
result.head()
```

## Aggregating tweets 
- Create any columns which may be needed for aggregation (month and year have been chosen as the grain for time) 
- Using pivot table function to explore average values for each lexical category for given date and location 

```python
result['date'] = pd.to_datetime(result['date'])
result['Period'] = result['date'].dt.to_period('M').dt.strftime('%m.%Y')
result.head()
```

```python
pd.pivot_table(result,index=[ "Period", "location"], values=list_cat,aggfunc=[np.mean])
```
